Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.

üåç Real-Time Use
Smart home automation systems

üì• Sample Input
Light Name: Bedroom Light
Action: ON

üì§ Expected Output
Bedroom Light is ON

```python
# Creating a class
class smart_light:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_status(self, action):
        self.status = action
        print(f"{self.name} is {self.status}")

light = smart_light("Bedroom Light")
light.set_status("ON")
     
```

Problem 2: Employee ID Card System
Create a class that stores employee name, ID, and department and displays ID card details.

Real-Time Use
Corporate HR systems

üì• Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT

üì§ Expected Output
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT


```python
# Creating a class
class employee_IDCard:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"

rahul_card = employee_IDCard(employee_name, employee_id, department)

rahul_card.display_card()
```

Problem 3: Mobile Contact Record
Create a class to store contact name and phone number and display contact information.

Real-Time Use
Mobile contact applications

üì• Sample Input
Contact Name: Anita
Phone Number: 9876543210

üì§ Expected Output
Contact Saved
Name: Anita
Phone: 9876543210



```python
# Creating a class
class contact_record:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


contact = contact_record("Anita", "9876543210")
contact.display_contact()
```

Problem 6: Delivery Address Manager

Customer Name: Suman
