Problem 1: Unique Words in a Sentence

Problem Statement:
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

Input
Enter sentence: python is easy and python is powerful

Expected Output
Unique words count: 5
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}

```python
Sentence="python is easy and python is powerful"
# splitting in words.
# getting the List of words and setting it as Set to get only unique words from sentence
words=set(Sentence.split())
print(f"Count of Unique Words: {len(words)}")
print(f"Unique Words: {words}")
```

Problem 2: Highest Salary from Employee Data

Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.

Example Input
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }

Expected Output
Highest Salary: Ravi - 75000

```python
# Dictionary ,Where Names are the Key and Salaries are the Value
employees = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }
Highest_salary=0
name_high_sal=""
#we will apply logic rather than max()
for emp_name , salary in employees.items(): #here items("Ravi":75000),emp_name="Ravi" and salary=75000 in first iteration of for loop
    if (salary > Highest_salary):
        Highest_salary=salary
        name_high_sal=emp_name
print(f"Highest Salary: {name_high_sal} - {Highest_salary}")
```

Problem 3: Find Maximum and Minimum Values

Problem Statement
Find the maximum and minimum values from a list of numbers.

Example Input
numbers = [45, 22, 89, 10, 66]

Expected Output
List: [45, 22, 89, 10, 66]
Max: 89
Min: 10

```python
numbers=[45,22,89,10,66]

max=numbers[0]
min=numbers[0]
for num in numbers:# it compares each num with 45
    if num>max:    # store the num greter than 45 and keep comparing ahead for higher values
        max=num
    if num<min:     # store the num smaller than 45 and keep comparing ahead for smaller values
        min=num
print(f"Max: {max} Min: {min}")
```

Problem 4: Count Products Above a Price Threshold

Problem Statement
Count how many products have a price greater than 1000.

Example Input
prices = [450, 1200, 899, 1500, 300]

Expected Output
Products above 1000: 2

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for p in prices:
    if p>1000:
        count+=1

print(f"Products above 1000: {count}")
```

Problem 5: Calculate Attendance Percentage

Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

Example Input
attendance = ["P", "P", "A", "P", "P"]

Expected Output
Attendance Percentage: 80.0

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
for a in attendance:
    if a=="P":
        count+=1 #gives "P" count
total=len(attendance)
percent=count/total*100
print(f"Attendance Percentage: {percent}")
```

Problem 6: Remove Duplicate Phone Numbers

Problem Statement
You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

Example Input
phone_numbers = [9876543210, 9123456789, 9876543210]

Expected Output
Unique phone numbers: {9876543210, 9123456789}

Problem 7: Count Character Frequency

Problem Statement
Count how many times each character appears in a given string.

Expected Output
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}

```python
text = "Pythonp"
# storing character count
count = {}
for char in text.lower(): #converts the characters in lower case
    # checking if character already in dict
    if char in count:
        count[char] += 1
    else:
        # adds new character
        count[char] = 1

print(f"Character frequency: {count}")
```
