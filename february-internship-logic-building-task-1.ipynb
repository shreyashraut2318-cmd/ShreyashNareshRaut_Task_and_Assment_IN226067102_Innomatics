```python
# 2.Pass/Fail Analyzer
# Problem Statement:
# Analyze student results.
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
Stud_Passed=[]
for m in marks:
    if (m>50):
        pass_count+=1
        Stud_Passed.append(m)
    else:
        fail_count+=1
print(f"Total no. of Students Passed: {pass_count}")
print(f"Total no. of Students Failed: {fail_count}")
print(f"Marks of Passed Students: {Stud_Passed}")
```

```python
# 3.Simple Data Cleaner
# Problem Statement:
# Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]
Cleaned_List=[]
for n in names:
    clean=n.strip().lower()
    Cleaned_List.append(clean)
print(f"Here is the list of Cleaned names: {Cleaned_List}")
```

```python
# 4.Message Length Analyzer
# Problem Statement:
# Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]
T_len=0
for msg in messages:
    length=len(msg)
    T_len+=length
    if(length>10):
        print(f"message: \"{msg}\" is longer than 10 characters")
    else:
        print(f"Message: {msg}, Length: {length}")
print(f"Total Length of whole message: {T_len}")
if (T_len>10):
    print("The total length of all messages exceeds 10 characters including spaces between words.")
```

```python
# 5. Error Message Detector
# Problem Statement:
# Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
Error_Count=0
for l in logs:
    if(l.lower()=="error"):
        Error_Count+=1
print(f"Their are total \"{Error_Count}\" ERROR Entries in logs")
```
