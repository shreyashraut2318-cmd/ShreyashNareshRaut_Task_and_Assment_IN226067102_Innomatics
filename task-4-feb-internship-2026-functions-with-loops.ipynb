```python
Problem 1: Mobile Recharge Validation System
Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps
Payment gateways
```

```python
# Function validating Mobile Recharge Amount 
def Mobile_recharge(Amount):
    # Check if recharge amount is less than or equal to minimum allowed
    if Amount<=50:
        return 'Recharge declined: Amount must be >=50'
        
    # Checking if amount matches any valid recharge plan
    elif Amount not in [199,299,399,599]:
        return 'Recharge declined: Enter a valid plan amount'

    # Recharge successful if every condition is met
    return 'Recharge Successful'

# Loop to allow user to retry until a valid recharge is entered
while True:
    #user input for recharge amount
    Amount = int(input('Enter the recharge amount:'))
    #calling the Mobile_recharge function
    result = Mobile_recharge(Amount)
    #display the result
    print(result)

    #Exit the loop if recharge is successful
    if result == 'Recharge Successful':
        break
```

Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise

Real-Time Use
Warehouse systems
Retail management


```python
# Creating a dictionary of products and its stock quantities
products = {}

# Number of products
n = int(input("Enter Number of Products: "))

# User input
for i in range(n):
    product = input(f"Enter product name{i+1}: ")
    stock = int(input("Enter stock quantity: "))
    products[product] = stock

# Function checking stock levels
def Check_Stock(products):
    # Iterating through each product inside dictionary
    for product, stock in products.items():
        #checking minimum stock quantity
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

# Function call
Check_Stock(products)
```

Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ≥ 50 else Fail

Real-Time Use
Online exam systems
Evaluation portals


```python
# Function to calculate result based on average marks
def result(marks):
    total = 0
    # Calculate total marks using loop
    for mark in marks:
        total += mark

    #calculating average marks
    avg = total/len(marks)

    #checking Pass or Fail
    if avg >= 50:
        return 'Pass'
    else:
        return 'Fail'

#User input for number of subjects
n = int(input('Enter the number of subjects:'))

#empty list to store marks
student_marks = []

#Taking user input for each subject marks
for i in range(n):
    m = int(input(f'Enter the marks of subject{i+1}: '))
    student_marks.append(m)

#Function call
evaluate = result(student_marks)
print(evaluate)
```

Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50
₹12 per km
Peak hour → 25% extra
Allow retry using while loop

Real-Time Use
Ride-hailing apps
Transport systems


```python
# Function Calculating Cab fare
def cab_fare_estimate(kms,peak_hour):
    base_fare = 50
    price_per_km = 12

    # Calculate base fare + distance fare
    total_fare = base_fare + kms * 12

    # Add 25% extra during peak hours
    if peak_hour == 'yes':
        total_fare += total_fare*0.25

    return f'Your total fare estimate is ₹{total_fare}'

# Retry loop
while True:
    #User input for number of kms and peak_hour insights
    kms = int(input('Enter the number of kms: '))
    peak_hour = input('Is it a peak hour? (yes/no) ').lower()

    #Function call
    estimate = cab_fare_estimate(kms,peak_hour)
    print(estimate)

    #Retry option
    retry = input('Do you want to retry? (yes/no) ').lower()
    if retry != 'yes':
        break
```

Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible

Real-Time Use
HR payroll systems
Internship tracking


```python
# Function to checking the attendance eligibility
def attendance_checker(attendance):
    p_count = 0
    
    # Counting number of present days
    for attend in attendance:
        if attend == 'P':
            p_count += 1

    # Calculate attendance percentage
    at_percentage = p_count/len(attendance)*100

    # Check eligibility
    if at_percentage >= 75:
        return 'Eligible'
    else:
        return 'Not Eligible'

#User input for number of days
n = int(input('Enter the number of days:'))

#Empty list to store attendance
attendance = []

#User input for each day
for i in range(n):
    while True:
        at = input(f'Enter the attendance on Day{i+1} (P or A): ').upper()
        if at in ['A','P']:
            attendance.append(at)
            break
        else:
            print('Invalid input for attendance')

#Function call
Eligibility = attendance_checker(attendance)
print(Eligibility)
```

Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)

Real-Time Use
Login systems
Security validation


```python
# Function Checking password strength
def password_checker(password):
    #checking minimum length
    if len(password) < 8:
        return 'Weak Password: minimum length must be 8 characters'

    has_digit = False
    has_special = False

    # Loop through each character in password
    for char in password: 
        if char.isdigit():
            has_digit= True
        if char in ['@','#','$']:
            has_special = True

    # Check digit condition
    if not has_digit:
        return 'Weak password: must contain atleast 1 digit'

    # Check special character condition
    if not has_special:
        return "Weak password: must contain atleast 1 special character (@#$)"
        
    # If all conditions are satisfied
    return "Approved! Strong Password"

#Retry until strong password is entered
while True:
    #user input for password
    password = input('Enter your password: ')

    #function calling
    check = password_checker(password)
    print(check)

    if check == 'Approved! Strong Password':
        break
```
